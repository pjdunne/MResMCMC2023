```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate a grid of x and y values
x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)
x, y = np.meshgrid(x, y)

# Define the parameters of the sinusoidal function
A = 1
B = 2
C = 0.65
D = 0

# Compute the z values
z = A * np.sin(B * (np.sqrt(x ** 2 + y ** 2) - C)) + D

# Create a 3D plot
fig = plt.figure(figsize=(10, 10)) # Set the figure size here (width, height)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
fig.colorbar(ax.plot_surface(x, y, z, cmap='viridis'))
# Add labels
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.set_title('3D Sine wave Function')
plt.savefig('3D Sine wave Function')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the double well potential
a = 1
b = 6

# Define the double well potential function
def V(x):
    return a*x**4 - b*x**2

# Create an array of x values from -3 to 3
x = np.linspace(-3, 3, 400)

# Create an array of y values using the function
y = V(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Double Well Potential')
plt.xlabel('x')
plt.ylabel('V(x)')
plt.grid(True)
plt.show()
```

The surface plot then represents the values of the potential function `V(x, y)` over the grid of x and y values.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants for the double well potential
a = 1
b = 6

# Define the double well potential function
def V(x, y):
    return a*x**4 - b*x**2 + y**2*0  # y**2*0 is added to make V a function of x and y, but y has no influence

# Create an array of x and y values from -3 to 3
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)

x, y = np.meshgrid(x, y)  # creates a rectangular grid for x and y

# Create an array of z values using the function
z = V(x, y)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Double Well Potential')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('V(x, y)')
plt.savefig('3D Double Well Function')
plt.show()
```

Holder Table Function known for its multiple local maxima and one global maximum, making it a good candidate for testing optimization algorithms.

f(x, y) = - abs(sin(x) * cos(y) * exp(abs(1 - sqrt(x^2 + y^2)/pi)))

where x and y are within the range [-10, 10].
It calculates the corresponding z values by passing x and y to the `holder_table` function. 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def holder_table(x, y):
    return -abs(np.sin(x) * np.cos(y) * np.exp(abs(1 - np.sqrt(x**2 + y**2) / np.pi)))

# Create an array of x and y values
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)

# Create a grid of x and y values
x, y = np.meshgrid(x, y)

# Calculate the z values
z = holder_table(x, y)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
ax.set_title('3D Plot of Holder Table Function')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
plt.savefig('3D Holder Table Function')
plt.show()
```

The Cross-in-Tray function is a mathematical function often used in optimization problems due to its multiple minima and maxima. 

f(x, y) = -0.0001 * (abs(sin(x)*sin(y)*exp(abs(100 - sqrt(x^2 + y^2)/pi))) + 1)^0.1

The function is evaluated in the square x ∈ [-10, 10], y ∈ [-10, 10].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def cross_in_tray(x, y):
    return -0.0001 * (np.abs(np.sin(x) * np.sin(y) * np.exp(np.abs(100 - np.sqrt(x**2 + y**2) / np.pi))) + 1)**0.1

# Create an array of x and y values
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)

# Create a grid of x and y values
x, y = np.meshgrid(x, y)

# Calculate the z values
z = cross_in_tray(x, y)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
ax.set_title('3D Plot of Cross-in-Tray Function')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
plt.savefig('3D Cross-in-Tray function')
plt.show()
```

The Eggholder function is mathematical function used in optimization problems. It's known for its numerous local minima, making it a challenging problem for optimization algorithms. 

The function is evaluated in the square x ∈ [-512, 512], y ∈ [-512, 512].

f(x, y) = -(y + 47) * sin(sqrt(abs(x/2 + (y+47)))) - x * sin(sqrt(abs(x - (y+47))))

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function
def eggholder(x, y):
    return -(y + 47) * np.sin(np.sqrt(abs(x/2 + (y + 47)))) - x * np.sin(np.sqrt(abs(x - (y + 47))))

# Create an array of x and y values
x = np.linspace(-512, 512, 400)
y = np.linspace(-512, 512, 400)

# Create a grid of x and y values
x, y = np.meshgrid(x, y)

# Calculate the z values
z = eggholder(x, y)

# Create the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='k')
ax.set_title('3D Plot of Eggholder Function')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(X, Y)')
plt.savefig('3D Eggholder Function')
plt.show()
```
